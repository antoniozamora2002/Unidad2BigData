```python
import matplotlib.pyplot as plt  # Importa la biblioteca Matplotlib.
import numpy as np  # Importa NumPy para operaciones matemáticas.

x = np.linspace(0, 10, 100)  # Crea 100 puntos entre 0 y 10 (inclusive).
y = np.sin(x)  # Calcula el seno de cada punto en `x`.

plt.figure(figsize=(8, 4))  # Crea una figura de 8x4 pulgadas.
plt.plot(x, y, label='sen(x)', color='blue', linestyle='-', linewidth=2)  # Grafica:
# `x`, `y`: Datos.
# `label`: Etiqueta para la leyenda.
# `color`: Color de la línea.
# `linestyle`: Estilo (continuo en este caso).
# `linewidth`: Grosor de la línea.

plt.title("Gráfico de Línea")  # Título del gráfico.
plt.xlabel("Eje X")  # Etiqueta del eje X.
plt.ylabel("Eje Y")  # Etiqueta del eje Y.
plt.grid(True)  # Activa la cuadrícula.
plt.legend()  # Muestra la leyenda.
plt.show()  # Renderiza el gráfico.
```

```python
categorias = ['Manzanas', 'Bananas', 'Cerezas', 'Dátiles']  # Nombres de las barras.
valores = [45, 30, 15, 10]  # Alturas de las barras.

plt.figure(figsize=(8, 4))  # Figura de 8x4 pulgadas.
plt.bar(categorias, valores, color=['red', 'yellow', 'pink', 'brown'])  # Crea barras:
# `categorias`: Eje X.
# `valores`: Eje Y.
# `color`: Lista de colores para cada barra.

plt.title("Gráfico de Barras")  # Título.
plt.xlabel("Frutas")  # Etiqueta eje X.
plt.ylabel("Cantidad")  # Etiqueta eje Y.
plt.show()  # Muestra el gráfico.
```

```python
x = np.random.rand(50)  # 50 números aleatorios entre 0 y 1.
y = x + np.random.normal(0, 0.1, 50)  # `y = x + ruido gaussiano`.

plt.figure(figsize=(8, 4))  # Tamaño de la figura.
plt.scatter(x, y, color='green', alpha=0.9, label='Datos')  # Puntos dispersos:
# `alpha`: Transparencia (0 a 1).
# `label`: Etiqueta para la leyenda.

plt.title("Gráfico de Dispersión")  # Título.
plt.xlabel("Variable X")  # Etiqueta eje X.
plt.ylabel("Variable Y")  # Etiqueta eje Y.
plt.legend()  # Muestra leyenda.
plt.show()  # Renderiza.
```

```python
datos = np.random.normal(0, 1, 1000)  # 1000 puntos de distribución normal (media=0, desv=1).

plt.figure(figsize=(8, 4))  # Tamaño de la figura.
plt.hist(datos, bins=30, color='purple', edgecolor='black')  # Histograma:
# `bins`: Número de barras (intervalos).
# `edgecolor`: Color del borde de las barras.

plt.title("Histograma")  # Título.
plt.xlabel("Valores")  # Etiqueta eje X.
plt.ylabel("Frecuencia")  # Etiqueta eje Y.
plt.show()  # Muestra.
```

```python
etiquetas = ['Python', 'Java', 'C++', 'JavaScript']  # Nombres de las porciones.
sizes = [45, 20, 15, 20]  # Tamaños de cada porción.
colores = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']  # Colores.

plt.figure(figsize=(6, 6))  # Figura cuadrada.
plt.pie(sizes, labels=etiquetas, colors=colores, autopct='%1.1f%%', shadow=True)  # Pastel:
# `autopct`: Muestra porcentajes con 1 decimal.
# `shadow`: Añade sombra.

plt.title("Lenguajes de Programación")  # Título.
plt.show()  # Renderiza.
```

```python
datos = [np.random.normal(0, std, 100) for std in range(1, 4)]  # 3 conjuntos de datos normales.

plt.figure(figsize=(8, 4))  # Tamaño de la figura.
plt.boxplot(datos, vert=True, patch_artist=True)  # Cajas:
# `vert`: Orientación vertical (True).
# `patch_artist`: Rellena las cajas con color.

plt.title("Gráfico de Caja")  # Título.
plt.xlabel("Grupos")  # Etiqueta eje X.
plt.ylabel("Valores")  # Etiqueta eje Y.
plt.show()  # Muestra.
```

```python
x = np.linspace(0, 10, 100)  # 100 puntos entre 0 y 10.

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))  # 1 fila, 2 columnas.
# `ax1`, `ax2`: Objetos de ejes para cada subplot.

ax1.plot(x, np.sin(x), 'r-', label='sen(x)')  # Subplot 1: Línea roja.
ax1.set_title("Gráfico de Línea")  # Título subplot 1.

ax2.scatter(x, np.cos(x), 'b.', label='cos(x)')  # Subplot 2: Puntos azules.
ax2.set_title("Gráfico de Dispersión")  # Título subplot 2.

plt.show()  # Renderiza ambos subplots.
```

```python
from mpl_toolkits.mplot3d import Axes3D  # Importa herramientas 3D.

fig = plt.figure(figsize=(8, 6))  # Crea figura.
ax = fig.add_subplot(111, projection='3d')  # Añade subplot 3D:
# `111`: 1 fila, 1 columna, primer gráfico.
# `projection='3d'`: Configuración 3D.

x = np.linspace(-5, 5, 100)  # Eje X.
y = np.linspace(-5, 5, 100)  # Eje Y.
X, Y = np.meshgrid(x, y)  # Crea una malla 2D para X e Y.
Z = np.sin(np.sqrt(X**2 + Y**2))  # Calcula Z (seno de la distancia al origen).

ax.plot_surface(X, Y, Z, cmap='viridis')  # Superficie 3D:
# `cmap`: Mapa de colores ('viridis').

ax.set_title("Superficie 3D")  # Título.
plt.show()  # Renderiza.
```

```python
import matplotlib.pyplot as plt

# Simulación de datos de una tienda online
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
ventas_mensuales = [1200, 1500, 1700, 1300, 1600, 1800]  # Ventas por mes en dólares

categorias = ['Electrónica', 'Ropa', 'Hogar', 'Juguetes']
ventas_por_categoria = [4500, 3000, 2000, 1500]  # Ventas acumuladas por categoría

visitas = [300, 400, 500, 350, 420, 550]  # Visitas al sitio por mes
satisfaccion = [3, 4, 5, 2, 4, 5, 5, 3, 4, 2, 5, 4]  # Encuestas de satisfacción (1 a 5)

ventas_diarias = [200, 220, 250, 190, 210, 230, 205, 240, 200, 195, 250, 210, 230]  # Ventas por día

# 1️⃣ GRÁFICO DE LÍNEAS - Ventas mensuales
plt.figure(figsize=(6, 4))  # Crear figura con tamaño 6x4 pulgadas
plt.plot(meses, ventas_mensuales, marker='o', linestyle='-', color='blue')  # Dibujar línea con puntos
plt.title('Ventas mensuales en USD')  # Título
plt.xlabel('Mes')  # Etiqueta eje x
plt.ylabel('Ventas')  # Etiqueta eje y
plt.grid(True)  # Mostrar rejilla
plt.tight_layout()  # Ajuste automático
plt.show()  # Mostrar gráfico

print("🔍 Análisis 1: Las ventas muestran una tendencia creciente general, con un pico en junio. Esto sugiere que las campañas de marketing o la estacionalidad están dando resultado.")

# 2️⃣ GRÁFICO DE BARRAS - Ventas por categoría
plt.figure(figsize=(6, 4))
plt.bar(categorias, ventas_por_categoria, color='orange')  # Barras verticales
plt.title('Ventas por categoría')
plt.xlabel('Categoría')
plt.ylabel('Ventas en USD')
plt.tight_layout()
plt.show()

print("🔍 Análisis 2: Electrónica es claramente la categoría líder en ventas, seguida por ropa. Esto indica una posible oportunidad para potenciar productos del hogar o juguetes.")

# 3️⃣ GRÁFICO DE DISPERSIÓN - Visitas vs Ventas
plt.figure(figsize=(6, 4))
plt.scatter(visitas, ventas_mensuales, color='green')  # Dibujar puntos
plt.title('Relación entre visitas y ventas')
plt.xlabel('Visitas al sitio')
plt.ylabel('Ventas en USD')
plt.grid(True)
plt.tight_layout()
plt.show()

print("🔍 Análisis 3: Existe una correlación positiva: a mayor número de visitas, mayores ventas. Esto valida la importancia de aumentar el tráfico al sitio web.")

# 4️⃣ HISTOGRAMA - Satisfacción del cliente
plt.figure(figsize=(6, 4))
plt.hist(satisfaccion, bins=5, color='purple', edgecolor='black')  # Histograma agrupado por niveles 1 a 5
plt.title('Distribución de satisfacción del cliente (1-5)')
plt.xlabel('Nivel de satisfacción')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()

print("🔍 Análisis 4: La mayoría de los clientes califican la experiencia con un 4 o 5, lo que indica una alta satisfacción. Sin embargo, hay algunos 2 que podrían investigarse.")

# 5️⃣ GRÁFICO DE TORTA - Proporciones por categoría
plt.figure(figsize=(6, 6))
plt.pie(ventas_por_categoria, labels=categorias, autopct='%1.1f%%', startangle=140)
plt.title('Proporción de ventas por categoría')
plt.tight_layout()
plt.show()

print("🔍 Análisis 5: El 45% de las ventas provienen de Electrónica. Visualmente se confirma que la mayoría de los ingresos se concentran en pocas categorías.")

# 6️⃣ BOXPLOT - Distribución de ventas diarias
plt.figure(figsize=(6, 4))
plt.boxplot(ventas_diarias, vert=True, patch_artist=True)  # Diagrama de caja para ver distribución
plt.title('Resumen estadístico de ventas diarias')
plt.ylabel('Ventas en USD')
plt.tight_layout()
plt.show()

print("🔍 Análisis 6: Las ventas diarias son estables, con valores centrados cerca de 220. No se observan valores atípicos extremos, lo que refleja buena consistencia operativa.")
```
