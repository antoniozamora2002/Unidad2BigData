```python
import matplotlib.pyplot as plt

# Simulaci√≥n de datos de una tienda online
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun']
ventas_mensuales = [1200, 1500, 1700, 1300, 1600, 1800]  # Ventas por mes en d√≥lares

categorias = ['Electr√≥nica', 'Ropa', 'Hogar', 'Juguetes']
ventas_por_categoria = [4500, 3000, 2000, 1500]  # Ventas acumuladas por categor√≠a

visitas = [300, 400, 500, 350, 420, 550]  # Visitas al sitio por mes
satisfaccion = [3, 4, 5, 2, 4, 5, 5, 3, 4, 2, 5, 4]  # Encuestas de satisfacci√≥n (1 a 5)

ventas_diarias = [200, 220, 250, 190, 210, 230, 205, 240, 200, 195, 250, 210, 230]  # Ventas por d√≠a
```

```python
# 1Ô∏è‚É£ GR√ÅFICO DE L√çNEAS - Ventas mensuales
plt.figure(figsize=(6, 4))  # Crear figura con tama√±o 6x4 pulgadas
plt.plot(meses, ventas_mensuales, marker='o', linestyle='-', color='blue')  # Dibujar l√≠nea con puntos
plt.title('Ventas mensuales en USD')  # T√≠tulo
plt.xlabel('Mes')  # Etiqueta eje x
plt.ylabel('Ventas')  # Etiqueta eje y
plt.grid(True)  # Mostrar rejilla
plt.tight_layout()  # Ajuste autom√°tico
plt.show()  # Mostrar gr√°fico

print("üîç An√°lisis 1: Las ventas muestran una tendencia creciente general, con un pico en junio. Esto sugiere que las campa√±as de marketing o la estacionalidad est√°n dando resultado.")
```

```python
# 2Ô∏è‚É£ GR√ÅFICO DE BARRAS - Ventas por categor√≠a
plt.figure(figsize=(6, 4))
plt.bar(categorias, ventas_por_categoria, color='orange')  # Barras verticales
plt.title('Ventas por categor√≠a')
plt.xlabel('Categor√≠a')
plt.ylabel('Ventas en USD')
plt.tight_layout()
plt.show()

print("üîç An√°lisis 2: Electr√≥nica es claramente la categor√≠a l√≠der en ventas, seguida por ropa. Esto indica una posible oportunidad para potenciar productos del hogar o juguetes.")
```

```python
# 3Ô∏è‚É£ GR√ÅFICO DE DISPERSI√ìN - Visitas vs Ventas
plt.figure(figsize=(6, 4))
plt.scatter(visitas, ventas_mensuales, color='green')  # Dibujar puntos
plt.title('Relaci√≥n entre visitas y ventas')
plt.xlabel('Visitas al sitio')
plt.ylabel('Ventas en USD')
plt.grid(True)
plt.tight_layout()
plt.show()

print("üîç An√°lisis 3: Existe una correlaci√≥n positiva: a mayor n√∫mero de visitas, mayores ventas. Esto valida la importancia de aumentar el tr√°fico al sitio web.")
```

```python
# 4Ô∏è‚É£ HISTOGRAMA - Satisfacci√≥n del cliente
plt.figure(figsize=(6, 4))
plt.hist(satisfaccion, bins=5, color='purple', edgecolor='black')  # Histograma agrupado por niveles 1 a 5
plt.title('Distribuci√≥n de satisfacci√≥n del cliente (1-5)')
plt.xlabel('Nivel de satisfacci√≥n')
plt.ylabel('Frecuencia')
plt.tight_layout()
plt.show()

print("üîç An√°lisis 4: La mayor√≠a de los clientes califican la experiencia con un 4 o 5, lo que indica una alta satisfacci√≥n. Sin embargo, hay algunos 2 que podr√≠an investigarse.")
```

```python
# 5Ô∏è‚É£ GR√ÅFICO DE TORTA - Proporciones por categor√≠a
plt.figure(figsize=(6, 6))
plt.pie(ventas_por_categoria, labels=categorias, autopct='%1.1f%%', startangle=140)
plt.title('Proporci√≥n de ventas por categor√≠a')
plt.tight_layout()
plt.show()

print("üîç An√°lisis 5: El 45% de las ventas provienen de Electr√≥nica. Visualmente se confirma que la mayor√≠a de los ingresos se concentran en pocas categor√≠as.")
```

```python
# 6Ô∏è‚É£ BOXPLOT - Distribuci√≥n de ventas diarias
plt.figure(figsize=(6, 4))
plt.boxplot(ventas_diarias, vert=True, patch_artist=True)  # Diagrama de caja para ver distribuci√≥n
plt.title('Resumen estad√≠stico de ventas diarias')
plt.ylabel('Ventas en USD')
plt.tight_layout()
plt.show()

print("üîç An√°lisis 6: Las ventas diarias son estables, con valores centrados cerca de 220. No se observan valores at√≠picos extremos, lo que refleja buena consistencia operativa.")
```
